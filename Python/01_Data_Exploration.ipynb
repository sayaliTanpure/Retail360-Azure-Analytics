### Cleaning Step 1: Datetime Conversion

Converted the following columns from string to datetime format:

- order_purchase_timestamp
- order_approved_at
- order_delivered_carrier_date
- order_delivered_customer_date
- order_estimated_delivery_date

**Reason:**
Datetime format is required for time-based analysis such as monthly sales, delivery duration, order trends, and business reporting.

```python
orders["delivery_days"] = (
    orders["order_delivered_customer_date"] - 
    orders["order_purchase_timestamp"]
).dt.days
```

```python
orders[[
    "order_delivered_customer_date",
    "order_purchase_timestamp",
    "delivery_days"
]].head()
```

```python
orders["purchase_year"] = orders["order_purchase_timestamp"].dt.year
```

```python
orders["purchase_month"] = orders["order_purchase_timestamp"].dt.month_name()
```

```python
orders["purchase_day"] = orders["order_purchase_timestamp"].dt.day
```

```python
orders["purchase_weekday"] = orders["order_purchase_timestamp"].dt.day_name()
```

```python
orders[[
    "order_purchase_timestamp",
    "purchase_year",
    "purchase_month",
    "purchase_day",
    "purchase_weekday"]].head()
```

```python
orders["is_weekend"] = orders["purchase_weekday"].isin(["Saturday","Sunday"])
```

```python
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"] -
    orders["order_estimated_delivery_date"]
).dt.days
```

```python
orders["is_delayed"] = orders["delivery_delay_days"] > 0
```

```python
orders[[
    "purchase_weekday",
    "is_weekend",
    "delivery_delay_days",
    "is_delayed"
]].head()
```

```python
orders["is_weekend"] = orders["is_weekend"].astype(int)
orders["is_delayed"] = orders["is_delayed"].astype(int)
```

```python
orders.head()
```

```python
products["product_weight_g"] = products["product_weight_g"].fillna(products["product_weight_g"].median())
```

```python
products["product_length_cm"] = products["product_length_cm"].fillna(products["product_length_cm"].median())
```

```python
products["product_height_cm"] = products["product_height_cm"].fillna(products["product_height_cm"].median())
```

```python
products["product_width_cm"] = products["product_width_cm"].fillna(products["product_width_cm"].median())
```

```python
products["product_category_name"] = products["product_category_name"].fillna("Unknown")
```

```python
products["product_volume_cm3"] = products["product_length_cm"] * products["product_height_cm"] * products["product_width_cm"]
```

```python
products["is_heavy"] = products["product_weight_g"] > 5000
```

```python
products["is_large"] = products["product_volume_cm3"] > 100000
```

```python
products["is_heavy"] = products["is_heavy"].astype(int)
products["is_large"] = products["is_large"].astype(int)
```
